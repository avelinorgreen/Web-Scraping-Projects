```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
# Function to sign in and go to foods page!

def login():
    
    url = "https://www.campbells.com.au/convenience/"
    username = '0032353419'
    password = 'Australia@1'

    driver.implicitly_wait(10)

    driver.get(url)

    body = driver.find_element(By.TAG_NAME,'body')

    body.send_keys(Keys.ESCAPE)

    login = driver.find_element(By.CLASS_NAME,'loginLink')

    login.click()

    user = driver.find_element(By.ID,'j_username')
    Pass = driver.find_element(By.ID,'j_password')

    user.send_keys(username)
    Pass.send_keys(password)

    button = driver.find_element(By.ID,'loginButton')

    button.click()
    
# Function to scrap the url's, brands, product names, and to see if they are in stock!   
    
def process_products(html):
    
    soup = BeautifulSoup(html, 'html.parser')
    products = soup.find_all("div",{"class":"span-5"})
    clean_product_list = []
    for product in products:
        data_dict = {}
        featured = product.find("div",{"class":"featuredProduct"})
        data_dict['product_url'] = featured.find('a')['href']
        data_dict['product_brand'] = featured.find("div",{"class":"brand"}).text.strip()
        data_dict['product_name'] = featured.find("div",{"class":"productName"}).text.strip()
        clean_product_list.append(data_dict)

        try:
            product.find("span",{"class":"stock"}).text
            data_dict['product_out_of stock'] = "Y"
        except:
            data_dict["product_out_of_stock"] = ""
        clean_product_list.append(data_dict)
        
    return clean_product_list
```

```python
# Using the functions!

driver = webdriver.Chrome('/Users/austinreyes/selenium_path/chromedriver')

login()

driver.get('https://www.campbells.com.au/convenience/foodservice?pageSize=100&q=%3Arelevance#')
html = driver.page_source
```

```python
process_products(html)
```
