```python
def convert_views(y_df):
    if 'K' in y_df['views']:
        views = float(y_df['views'].split('K')[0])*1000
        return views
    elif 'M' in y_df['views']:
        views = float(y_df['views'].split('M')[0])*1000000
        return views
```

```python
df['clean_views'] = df.apply(convert_views,axis=1)
```
